```python
# 4a. Made Changes to Dataset - Version 5 - Converting Non-numerical Variables In Excel And Python
import pandas as pd
v5 = pd.read_csv("image-descriptions-DFE5.csv", header = 0)
```

```python
# 4b. Checking The Variable Types
v5.dtypes
```

```python
# 4c. Converting Image and Phrase Variable As Category Variables
v5["Image"] = v5["Image"].astype('category')
v5["Phrase"] = v5["Phrase"].astype('category')
v5.dtypes
```

```python
# 4d. Adding New Variables
v5["Image New"] = v5["Image"].cat.codes
v5["Phrase New"] = v5["Phrase"].cat.codes
v5.dtypes
```

```python
# 4e. Final Version Of Dataset - Version 5 
v5
```

```python
# 5a. Sorted The Data To See If The Image And Phrase Variables Were Converted Correctly
import pandas as pd
v6 = pd.read_csv("image-descriptions-DFE6.csv", header = 0)
v6
```

```python
# 5b. Converting Variables And Adding New Variables
v6["Image"] = v6["Image"].astype('category')
v6["Phrase"] = v6["Phrase"].astype('category')
v6["Image New"] = v6["Image"].cat.codes
v6["Phrase New"] = v6["Phrase"].cat.codes
v6.dtypes
```

```python
# 5c. Final Version Dataset - Version 6
v6
```

```python
# 6. Multiple Linear Regression Model - Use Version 5 to Run Model
import pandas as pd
from sklearn import linear_model
import statsmodels.api as sm

X1 = v5[['Trusted Judgement','Response New','Confidence Score New','Image New','Phrase New']] 
Y1 = v5['Unit Id']

regr = linear_model.LinearRegression()
regr.fit(X1, Y1)

print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)

X1 = sm.add_constant(X1) # adding a constant
 
model1 = sm.OLS(Y1, X1).fit()
predictions = model1.predict(X1) 
 
print_model1 = model1.summary()
print(print_model1)
```
